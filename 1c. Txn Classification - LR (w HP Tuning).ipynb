**Classification of Transactions using Logistic Regression**

--------------
Key insights so far:
- De-duplicated address table and enhanced transaction table have been created with the ending '_clean'.
- 10 wallet addresses and transactions have been validated using BTC scan, meaning the address and transaction data in the Elliptic++ dataset match the public blockchain leger. As a result, the Elliptic++ dataset can be used for future analysis.
- Key behavioural patterns between illicit and licit addresses have been explored and identified.
- Key behavioural patterns between illicit and licit transactions have been explored and identified.
- Validation of address classes has been undertaken by manually calculating the address class based on address logic using global transaction ratio from the Elliptic++ paper. There is a small discrepancy of 214 addresses which have a mismatched label (unknown instead of licit), representing 0.03% of the address population. The author has been contacted for further information regarding the address class mismatch and the rationale for global address conditions.
- The number of transactions in each timestamp are all connected, so each timestamp is a background graph.
- There are no transactions as part of the background graph which are not connected to other transactions.


Purpose of this notebook:
- The purpose of this notebook is to apply a logistic regression model to transactions to make a prediction on whether 'unknown' class transactions are licit or illicit. After all transactions are labelled, addresses will be able to be labelled as Licit or Illicit based on the proportion of licit v illicit transactions each address is linked to.

Measure of success:
- All unknown transaction classes are labelled as licit or illicit based on the output of the logistic regression model.
- Accuracy > 85%, Recall >75% from validation data.

Outcome:
- All 157,205 unknown transaction classes are labelled as licit or illicit. 119,618 unknown txn (76%) are now Licit and 	37,587 unknown txn (24%) are now Illicit.
- Recall for illicit txn is 79%, accuracy is 94%, gini = 0.7417.

##### Training and Validation using Cross-Fold on Temporal Data
-------------------

```python
drop_training_cols = ['txId', 'Time step', 'class', 'class_label', 'data_split', 'training_label', 'pred_proba']

columns_to_normalize = [
    col for col in dataset.columns
    if col not in drop_training_cols and not col.startswith("Local_feature_") and not col.startswith("Aggregate_feature_")
    ]

print(columns_to_normalize)
```

```python
# --- Cross-Fold Validation Parameters (unchanged) ---
timestep_field = "Time step"
data_split_field = "data_split"
valid_splits = ['Train/Val']
obs_size = 30000
train_ratio = 0.5
threshold = 0.5
num_folds = 10
```

```python
from sklearn.experimental import enable_halving_search_cv  # noqa: F401
from sklearn.model_selection import HalvingGridSearchCV, TimeSeriesSplit
from sklearn.linear_model import LogisticRegression

tscv = TimeSeriesSplit(n_splits=3)

halving_lr = HalvingGridSearchCV(
    estimator=LogisticRegression(
        solver="saga",
        penalty="elasticnet",
        class_weight="balanced",
        max_iter=200,              # start small; scheduler increases it
        random_state=42
    ),
    param_grid={
        "C": [0.03, 0.1, 0.3, 1],  # tiny grid
        "l1_ratio": [0.0, 0.5, 1.0]
    },
    resource="max_iter",
    min_resources=200,
    max_resources=2000,
    factor=2,
    scoring="recall",
    cv=tscv,
    n_jobs=-1,
    verbose=1,
    refit=True
)
```

```python
summary_df, trained_models, scalers, fold_indices_dict = build_model.run_cross_validation(
    input_df=dataset_imputed,
    timestep_field=timestep_field,
    drop_training_cols=drop_training_cols,
    columns_to_normalize=columns_to_normalize,
    data_split_field=data_split_field,
    valid_splits=valid_splits,
    model=halving_lr,
    obs_size=obs_size,
    train_ratio=train_ratio,
    threshold=threshold,
    num_folds=num_folds
)
```

```python
import numpy as np
import pandas as pd
from sklearn.inspection import permutation_importance

# === set this explicitly to the fold your function printed ===
selected_fold = "Fold 6"   # e.g., "Fold 4" based on your Top Models Ranked

# Use your already-selected objects
est = best_model.best_estimator_ if hasattr(best_model, "best_estimator_") else best_model
scaler = best_scaler

# ---- Build the OUTER-VALIDATION slice for that fold (explicit, no auto-detect) ----
val_idx = fold_indices_dict[selected_fold]["val_indices"]
Xv = dataset_imputed.loc[val_idx, est.feature_names_in_].copy()
yv = dataset_imputed.loc[val_idx, "class"].astype(int).values

# Apply the same scaling used during CV (only to the normalized columns)
norm_cols = [c for c in Xv.columns if c in columns_to_normalize]
if norm_cols and scaler is not None:
    Xv.loc[:, norm_cols] = scaler.transform(Xv[norm_cols])

# ---- Permutation importance (choose your metric) ----
perm_metric = "f1"   # or "recall" if recall is your focus
perm = permutation_importance(
    est, Xv, yv,
    n_repeats=10,
    scoring=perm_metric,
    n_jobs=-1,
    random_state=42
)

perm_df = (
    pd.DataFrame({
        "feature": est.feature_names_in_,
        "perm_importance_mean": perm.importances_mean,
        "perm_importance_std": perm.importances_std
    })
    .sort_values("perm_importance_mean", ascending=False)
    .reset_index(drop=True)
)

# Optional: normalise to 0–100 for nicer presentation
top = perm_df["perm_importance_mean"].max()
if top and top > 0:
    perm_df["perm_importance_index_0_100"] = (100 * perm_df["perm_importance_mean"] / top).round(1)

# ---- Gini (impurity) importances from the RF ----
gini_df = (
    pd.DataFrame({
        "feature": est.feature_names_in_,
        "gini_importance": getattr(est, "feature_importances_", np.zeros(len(est.feature_names_in_)))
    })
    .sort_values("gini_importance", ascending=False)
    .reset_index(drop=True)
)

# ---- Combined (nice for thesis) ----
imp_merged = (
    perm_df[["feature","perm_importance_mean","perm_importance_std"]]
    .merge(gini_df, on="feature", how="outer")
    .assign(
        rank_perm=lambda d: d["perm_importance_mean"].rank(ascending=False, method="min"),
        rank_gini=lambda d: d["gini_importance"].rank(ascending=False, method="min")
    )
    .sort_values("rank_perm")
    .reset_index(drop=True)
)

# Peek
display(perm_df.head(20))
display(gini_df.head(20))
display(imp_merged.head(25))
```
